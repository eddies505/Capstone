# Number based dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_metrics(model_names, accuracies, precisions, recalls, f1scores):
    plt.figure(figsize=(12, 8))

    # Accuracy Plot
    plt.subplot(2, 2, 1)
    sns.barplot(x=model_names, y=accuracies, palette='viridis')
    plt.title('Accuracy')

    # Precision Plot
    plt.subplot(2, 2, 2)
    sns.barplot(x=model_names, y=precisions, palette='magma')
    plt.title('Precision')

    # Recall Plot
    plt.subplot(2, 2, 3)
    sns.barplot(x=model_names, y=recalls, palette='plasma')
    plt.title('Recall')

    # F1 Score Plot
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    sns.barplot(x=model_names, y=f1scores, palette='cividis')
    plt.title('F1 Score')

    plt.tight_layout()
    plt.show()

# Sample data
#model_names = ['Kaplan-Meier Survival', 'Cox Model']
# Corrected sample data
model_names = ['Kaplan-Meier Survival', 'Cox Model', 'ResNet']
accuracies = [0.02, 0.01, 0.02]
precisions = [0.01, 0.01, 0.01]
recalls = [0.01, 0.01, 0.01]
f1scores = [0.01, 0.01, 0.01]


# Plot the metrics
# Plot the metrics
plot_metrics(model_names, accuracies, precisions, recalls, f1scores)
```
